## Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Frogs', 'Dogs', 'Cats', 'Birds']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, 
        explode=explode, 
        labels=labels, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=90)
ax1.axis('equal')
```

```python
labels = ['Frogs', 'Dogs', 'Cats', 'Birds']
sizes = [15, 30, 45, 20]
explode = (0, 0.2, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, 
        explode=explode, 
        labels=labels, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=90)
ax1.axis('equal');
```

## Functions

```python
def flip_coin() -> str:
  flip = np.random.randint(2)
  if flip == 0:
    return "heads"
  elif flip == 1:
    return "tails"
  else:
    return "wtf"
```

```python
[print(f"The coin landed {flip_coin()}") for i in range(10)];
```
