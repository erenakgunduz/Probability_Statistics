```python
t = np.arange(0.0, 2.0, 0.01) # Step size of 0.01
s = np.sin(2*np.pi*t)

plt.plot(t, s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20) # Strings with r enclosed with dollar signs to use LaTeX in the text
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20) # First we define coordinates for the text location on the plot
plt.text(0.5, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',fontsize=15)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

```python
# Scatter plot on polar axis
# https://matplotlib.org/stable/gallery/pie_and_polar_charts/polar_scatter.html
# Fixing random state for reproducibility
np.random.seed(19680801) # as they said - fixes the random values so the output is always exactly the same as their example

# Compute areas and colors
N = 150 # The number of values we'll be viewing
r = 2 * np.random.rand(N) # The actual array of values in question - let's say we get our "radiuses" (radii?) by multiplying our array containing 150 values at random between 0 and 1 by 2, so now they'll go up to 2
theta = 2 * np.pi * np.random.rand(N) # Using the circumference formula to calculate the angle at which a value will appear on the polar plot
# Why do I say "a" value and not "the" value? Why don't we just substitute the 2 * np.random.rand(N) here with r itself?
# Because then the scatter plot would no longer be very randomized - that would instead generate a uniform spiral pattern of ascending radiuses
area = 200 * r**2 # Area formula but cranked up by a scale factor of about 64 for much better visibility (200/np.pi)
colors = theta # The angle at which the point appears will map directly to its color

fig = plt.figure() # Simply generating the figure, nothing more specified just yet
ax = fig.add_subplot(projection='polar') # We add the first subplot and make it a polar projection
c = ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.75) # To the "ax" polar projection we pass in the necessary arguments for the scatter plot
# The angles correspond to the x values (map directly to color) and the radiuses are the output/y values (scale directly to size in plot), passing in the color system we defined and using the hue, saturation, value colormap
# With an alpha blending (transparency of the dots) of 75% opacity
```
